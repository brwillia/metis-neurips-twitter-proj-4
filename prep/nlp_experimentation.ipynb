```python
import pandas as pd
import numpy as np

f = open('jkr_hp1.txt')
raw = f.read()
```

```python
import re
import string

def alphanumeric(in_str):
    return re.sub('\w*\d\w*', ' ', in_str)

def punc_lower(in_str):
    return re.sub('[%s]' % re.escape(string.punctuation), ' ', in_str.lower())

hp1 = ' '.join(punc_lower(alphanumeric(raw)).split())
```
